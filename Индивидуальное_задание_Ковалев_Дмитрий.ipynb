## 1.2. Удаление "лишних" столбцов

Проведя первичный анализ, я пришел в к выводу, что удалить следует столбцы под названием:
* __has_ir_blaster__ -- так как это функция очень узконаправленна и не требуется большинству людей
* __processor_brand__ -- обычно процессоры создают такие компании, как snapdragon и helio, но обычно у несведующего человека названия этих брендов не имеют большого смысла.
* __num_cores__ -- так как у телефонов в среднем большое количество ядер(6 или 8), дабы уместить хорошую производительность в маленькое устройство
* __fast_charging_available__ -- эта функция лишь немного облегчает пользование смартфоном, и не имеет кардинальное влияние на выбор телефона
* __fast_charging__ -- невнятный столб, который как бы показывает, за сколько минут полностью заряжается аккумулятор с помощью быстрой зарядки, хотя это можно проверить самостоятельно.
* __num_rear_cameras__ -- не представляет труда посчитать визуально, сколько их на самом деле.
* __num_front_cameras__ -- обычно не больше 1, однако даже если больше, большого влияния на съемку это не понесет.
* __extended_memory_available__ -- обычно встроенной памяти всегда достаточно для комфортного пользования моделью.
* __extended_upto__ -- непонятно, что автор датасета имеет в виду, вводя этот столб **(не ясно его назначение)**
* __resolution_width__ -- чаще всего показатель диагонали экрана имеет большее значение, нежели чем этот показатель.
* __resolution_height__ -- чаще всего показатель диагонали экрана имеет большее значение, нежели чем этот показатель.

**Итого, я посчитал нужным удалить 11 столбцов за ненужностью содержащейся в них информации.**

```python
df.drop(labels = 'has_ir_blaster', axis = 1, inplace = True)
df.drop(labels = 'processor_brand', axis = 1, inplace = True)
df.drop(labels = 'num_cores', axis = 1, inplace = True)
df.drop(labels = 'fast_charging_available', axis = 1, inplace = True)
df.drop(labels = 'fast_charging', axis = 1, inplace = True)
df.drop(labels = 'num_rear_cameras', axis = 1, inplace = True)
df.drop(labels = 'num_front_cameras', axis = 1, inplace = True)
df.drop(labels = 'extended_memory_available', axis = 1, inplace = True)
df.drop(labels = 'extended_upto', axis = 1, inplace = True)
df.drop(labels = 'resolution_width', axis = 1, inplace = True)
df.drop(labels = 'resolution_height', axis = 1, inplace = True)
```

## 3. Фильтрация данных

```python
#метод query
# 1. Выбрать процессоры с высокой частотой (более 3.0 ГГц)
processors_with_high_speed = df.query('(processor_speed) > 3.0')[['brand_name', 'model', 'price', 'rating', 'processor_speed']]
processors_with_high_speed
```

```python
# метод where
# Отбор телефонов Apple с рыночной стоимостью более 70 000
top_apple_phones = df.where((df['brand_name'] == "apple") & (df['price'] > 70000))
top_apple_phones = top_apple_phones.dropna()
top_apple_phones
```

## 3. "Портрет" типичного потребителя (вывод к разделу 3)

На основе анализа данных можно выделить следующие характеристики типичной модели на рынке смартфонов:

__Объем Аккумулятора:__

* Средний объем: около 4-6 тыс. МаЧ. Это объем, при котором производство аккумуляторов и использование их людьми достигает баланса для обоих сторон.

__Объем ОЗУ:__

* Средний объем: Это либо 6, либо 8, или же 12 Гигабайт оперативной памяти. Интересный факт, что самая дорогая модель на рынке имеет объем ОЗУ всего лишь 2 Гб.
* Также можно подметить, что телефоны на IOS не обладают объемом ОЗУ, выше чем 8 гб, в то время как android и другие системы могут доходить до объемов в 18 Гб в модели.
* В заключение можно отметить, что разница между 6 и 12 Гб в цене не такая уже большая, а, сдедовательно, объем ОЗУ не является определяющим фактором при составлении цены телефона.
